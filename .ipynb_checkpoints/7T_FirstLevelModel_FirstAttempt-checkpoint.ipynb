```python
%matplotlib inline

import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltools.stats import regress, zscore
from nltools.data import Brain_Data, Design_Matrix
from nltools.stats import find_spikes 
from nilearn.plotting import view_img, glass_brain, plot_stat_map
```

```python
single_subject = "/Users/phoenix/Desktop/imaging_studies/AEV/Data/Behavioral_Data/AffVids_logfile_101CorrectedWithVideoNames.txt"
single_subject_brain = '/Users/phoenix/Desktop/imaging_studies/AEV/Data/AEV/BIDS_fmriprep/fmriprep/ses-01/sub-101/func/sub-101_task-AEV_run-01_bold_space-MNI152NLin2009cAsym_variant-smoothAROMAnonaggr_preproc.nii.gz'
subject_brain = Brain_Data(single_subject_brain)
subject_behavioural=pd.read_csv(single_subject,sep=' ', skiprows=1, header=None)
```

```python
first_run = subject_behavioural.iloc[0:12,[1,9,10]]
TR = .001
raw_dm = np.zeros((205*2340,3))
for index, row in first_run.iterrows():
    #for each row get onset time
    
    onset = round(row.iloc[1]*1000)
    offset = round(row.iloc[1]*1000) + 5 * 1000
    condition = int(row.iloc[0])-1
    raw_dm[onset:offset,condition]=1
```

```python
dm = Design_Matrix(raw_dm, sampling_freq=1./TR, columns=['heights','pain','social'])
dm = dm.convolve()
dm = dm.iloc[::2340,:]
dm.sampling_freq= 1/2.34
dm = dm.reset_index(drop=True)
```

```python
nuisance_regressors_path = '/Users/phoenix/Desktop/imaging_studies/AEV/Data/AEV/BIDS_fmriprep/fmriprep/ses-01/sub-101/func/sub-101_task-AEV_run-02_bold_confounds.tsv'
confounds = pd.read_csv(nuisance_regressors_path, sep='\t')
confounds = confounds[['Cosine00', 'Cosine01', 'Cosine02', 'Cosine03', 'Cosine04', 'Cosine05',
       'NonSteadyStateOutlier00', 'X', 'Y', 'Z', 'RotX', 'RotY', 'RotZ']]

cov = Design_Matrix(confounds, sampling_freq = 1/2.34)

cov = cov.fillna(0)

# Retain a list of nuisance covariates (e.g. motion and spikes) which we'll also want to also keep separate for each run
cov_columns = cov.columns
cov = cov.add_poly(1).add_dct_basis()
```

```python
full = pd.concat([dm,cov], axis=1)
```

```python
subject_brain.X = full
results = subject_brain.regress()
```
